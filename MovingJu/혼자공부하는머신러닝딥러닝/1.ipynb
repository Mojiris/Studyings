### 확인 문제

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```

## 전체 데이터 구조 확인

```python
import numpy as np

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

print(f"shape : {input_arr.shape}")

input_arr
```

## 데이터 셔플링

```python
np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index)

index
```

```python
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
# 혹은

import numpy as np
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state=42, test_size=0.25)

train_input = np.array(train_input)
test_input = np.array(test_input)
train_target = np.array(train_target)
test_target = np.array(test_target)
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

## 지도 학습 수행

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

## Test the model

```python
kn.predict(test_input)
```

```python
test_target
```

## 데이터 Normalize

```python
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)

mean, std
```

```python
train_scaled = (train_input - mean) / std
test_scaled = (test_input - mean) / std
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(test_scaled[:, 0], test_scaled[:, 1])
plt.show()
```

```python
sus_fish = np.array([25, 150])
sus_fish_normalized = (sus_fish - mean) / std
```

```python
kn = KNeighborsClassifier()

kn.fit(train_scaled, train_target)
```

```python
kn.score(test_scaled, test_target)
```

```python
kn.predict([sus_fish_normalized])
```

```python
distances, indexes = kn.kneighbors([sus_fish_normalized])
plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(sus_fish_normalized[0], sus_fish_normalized[1],marker="^")
plt.scatter(
    train_scaled[indexes, 0],
    train_scaled[indexes, 1],
    marker="D"
)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```
