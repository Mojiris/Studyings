```python
import kagglehub

path = kagglehub.dataset_download("vipullrathod/fish-market")
```

```python
import pandas as pd

data = pd.read_csv(f"{path}/Fish.csv")

data
```

```python
import matplotlib.pyplot as plt

bream_length, bream_weight = [], []
for idx, val in enumerate(data.iloc[:, 0]):
  if val != "Bream":
    continue
  bream_length.append(data.iloc[idx, 2])
  bream_weight.append(data.iloc[idx, 1])



plt.scatter(bream_length, bream_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
import matplotlib.pyplot as plt

bream_length, bream_weight = [], []
smelt_length, smelt_weight = [], []
for idx, val in enumerate(data.iloc[:, 0]):
  if val == "Bream":
    bream_length.append(data.iloc[idx, 2])
    bream_weight.append(data.iloc[idx, 1])
    continue
  if val == "Smelt":
    smelt_length.append(data.iloc[idx, 2])
    smelt_weight.append(data.iloc[idx, 2])


plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
import matplotlib.pyplot as plt

bream_length = data[data["Species"] == "Bream"].iloc[:, 2]
bream_weight = data[data["Species"] == "Bream"].iloc[:, 1]

plt.scatter(bream_length, bream_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```
